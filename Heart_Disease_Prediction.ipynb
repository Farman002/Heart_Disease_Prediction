```python
X=df.iloc[:,:-1].values
Y=df.iloc[:,-1].values
df
```

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,Y_train,Y_test=tts(X,Y,test_size=0.2,random_state=0)
X_test,Y_test
```

```python
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
X_train=sc_X.fit_transform(X_train)
X_test=sc_X.fit_transform(X_test)
X_train
```

```python
from sklearn.svm import SVC
classifier=SVC(kernel='linear',random_state=0)
classifier.fit(X_train,Y_train)
```

```python
Y_pred = classifier.predict(X_test)
Y_pred
```

```python
from sklearn import metrics
print("Accuracy for SVM prediction:",metrics.accuracy_score(Y_test, Y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
decisiontreeclassifier=DecisionTreeClassifier(criterion='entropy',random_state=0)
decisiontreeclassifier.fit(X_train,Y_train)
```

```python
Y_pred = decisiontreeclassifier.predict(X_test)
Y_pred
```

```python
from sklearn import metrics
print("Accuracy for decision tree prediction:",metrics.accuracy_score(Y_test, Y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB
nbclassifier=GaussianNB()
nbclassifier.fit(X_train,Y_train)
```

```python
Y_pred = nbclassifier.predict(X_test)
Y_pred
```

```python
from sklearn import metrics
print("Accuracy for Naive Bayes prediction:",metrics.accuracy_score(Y_test, Y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
classifier.fit(X_train,Y_train)
```

```python
Y_pred = classifier.predict(X_test)
Y_pred
```

```python
from sklearn import metrics
print("Accuracy for Random forest prediction:",metrics.accuracy_score(Y_test, Y_pred))
```
